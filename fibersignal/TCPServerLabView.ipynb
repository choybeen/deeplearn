```python
import socket
import threading
import numpy as np
import struct
import matplotlib.pyplot as plt
import time
import random

smpnum = 8
pout = 4
dtrLen = smpnum*3000
dtsLen = smpnum*pout + 3
rcvLast = hrtLast = time.time()
reCon = 30  # reconnect time
rdata = np.arange(0, dtrLen/10, 0.1)
uprdata = np.zeros(dtrLen)
nloop = range(0,dtrLen)
strrdata = ''.join([str(x) for x in rdata])
print('rdata: ', len(rdata), ',', rdata[:10])
sdata = np.arange(0, dtsLen*0.1, 0.1)*1.1+1
sdata[0] = smpnum*pout
sdata[1] = smpnum
sdata[2] = pout
upsdata = sdata
print('sdata: ', len(sdata), ',', sdata[:10])
```

```python
#处理客户端请求
def tcpserver(sock,addr):
    print('Accept new connection from %s:%s...' % addr)
    count=0
    while count<19:
        count += 1
        sock.sendall(rdata)
        time.sleep(0.001)
        recv_data = sock.recv(dtrLen)
        rcv_len = len(recv_data)
        raw_leng, rows, cols = struct.unpack('3i', recv_data)[0:12]
        print(count, " : ", rcv_len, raw_leng, rows, cols)            
        unpackdata = struct.unpack('6f', recv_data[12:36])
        print(unpackdata[0:24])
        time.sleep(0.001)
    sock.close()
    print('Connection from %s:%s closed' % addr)

ssvr=socket.socket(socket.AF_INET,socket.SOCK_STREAM)
ssvr.bind(('localhost',6310))#绑定地址和端口
ssvr.listen(20)
print('服务器等待连接.....')
svr,adds=ssvr.accept()
print ("客户端已连接—> \n",adds)
#创建新线程来处理每个客户端连接
#target=tcplink代表新线程要运行哪个函数
#args=(sock,addr)代表向这个方法传的参数
ts=threading.Thread(target=tcpserver,args=(svr,adds))
ts.start()  #启动这个线程
```
